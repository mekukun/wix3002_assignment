```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
posts1 = pd.read_csv('./data/political_figs/IsmailSabri60_posts.csv')
posts1['post_split'] = posts1['Post'].apply(lambda x: x.split(" ") if isinstance(x, str) else x)
posts1['tokens_num'] = posts1['post_split'].apply(lambda x: len(x))

posts2 = pd.read_csv('./data/political_figs/DrZahidHamidi_posts.csv')
posts2['post_split'] = posts2['Post'].apply(lambda x: x.split(" ") if isinstance(x, str) else x)
posts2['tokens_num'] = posts2['post_split'].apply(lambda x: len(x))
```

```python
x1, y1 = np.unique(posts1['tokens_num'], return_counts=True)
x2, y2 = np.unique(posts2['tokens_num'], return_counts=True)
(x1,y1), (x2,y2)
```

```python
# Calculate the probabilities for each set
prob1 = y1 / np.sum(y1)
prob2 = y2 / np.sum(y2)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
ax1.bar(x1, prob1, edgecolor='black', linewidth=0.5, width=1, color='orange')
ax2.bar(x2, prob2, edgecolor='black', linewidth=0.5, width=1, color='yellow')
# for i, v in zip(x1, y1):
#     ax1.text(i, v + 0.5, str(v), ha='center', fontsize=6)

highest_x = np.max([np.max(x1), np.max(x2)])
highest_y = np.max([np.max(prob1), np.max(prob2)])

ax1.set_xlabel("Number of Tokens")
ax1.set_ylabel("Probability Distribution")
ax1.set_title("Token Histogram")
ax1.set_xlim(0, highest_x)
ax1.grid()

ax2.set_xlabel("Number of Tokens")
ax2.set_ylabel("Probability Distribution")
ax2.set_title("Token Histogram")
ax2.set_xlim(0, highest_x)
ax2.grid()

fig.tight_layout()
```
